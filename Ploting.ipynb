```python
data=pd.read_csv(r"C:\Users\visha\OneDrive\market_data.csv",usecols=('Item_Type','Outlet_Size','Item_Outlet_Sales'))
```

```python
data.isnull().mean()*100
```

```python
X=data
Y=data['Item_Outlet_Sales']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train['Item_Type_impute']=X_train['Item_Type']
X_test['Item_Type_impute']=X_test['Item_Type']

X_train['Outlet_Size_impute']=X_train['Outlet_Size']
X_test['Outlet_Size_impute']=X_test['Outlet_Size']
```

```python
X_train['Item_Type_impute'][X_train['Item_Type_impute'].isnull()]=X_train['Item_Type'].dropna().sample(X_train['Item_Type'].isnull().sum()).values
X_test['Item_Type_impute'][X_test['Item_Type_impute'].isnull()]=X_train['Item_Type'].dropna().sample(X_test['Item_Type'].isnull().sum()).values
X_train['Outlet_Size_impute'][X_train['Outlet_Size_impute'].isnull()]=X_train['Outlet_Size'].dropna().sample(X_train['Outlet_Size'].isnull().sum()).values
X_test['Item_Type_impute'][X_test['Item_Type_impute'].isnull()]=X_train['Outlet_Size'].dropna().sample(X_test['Outlet_Size'].isnull().sum()).values

```
