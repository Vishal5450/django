```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv(r"C:\Users\visha\OneDrive\placement.csv")
dataset.head()
dataset.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="scor",data=dataset,hue="placement")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_trian,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel='linear')
sv.fit(x_trian,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_trian,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
